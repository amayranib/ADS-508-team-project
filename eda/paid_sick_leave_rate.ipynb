## Find information on Paid sick leave

```python
jeda.list_unique_values(df, "Provision")
```

## Create dataframe for "Paid sick leave"

```python
df_provision = df[df["Provision"].str.contains("Paid sick leave", case=False, na=False)]
df_provision.head()
```

```python
df_provision["Provision"].unique()
```

## Filter BLS for "All workers"

```python
filtered_bls = df_provision[df["Characteristic category"] == "All workers"]

display(filtered_bls.head())
display(filtered_bls.shape)
```

```python
selected_columns = [
    "Year",
    "Provision",
    "Industry",
    "Industry code",
    "Estimate",
    "Estimate category",
]

bls_summary = filtered_bls[selected_columns].copy()

display(bls_summary.head())
display(bls_summary.shape)
```

```python
jeda.long_report(bls_summary)
```

```python
for cols in selected_columns:
    jeda.list_unique_values(bls_summary, cols)
```

## Get Industry Year Avg of Workers that have access to Paid Sick Leave

### Final Data Dictionary
| Field         | Field Description                                  |
|---------------|----------------------------------------------------|
| Year          | Reference year.                                    |
| Industry      | Description of the industry code.                  |
| Industry code | 6-digit NAICS code identifying industry.           |
| Avg_Estimate  | Estimate of employees with access to Paid Leave Benefits |


```python
industry_year_avg = (
    bls_summary.groupby(["Year", "Industry", "Industry code"])["Estimate"]
    .mean()
    .reset_index()
    .rename(columns={"Estimate": "Avg_Estimate"})
)

display(industry_year_avg.head())
display(industry_year_avg.shape)
```

```python
industry_year_avg.to_csv("industry_year_avg.csv", index=False)
```
